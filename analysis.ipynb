YTD TOTAL DISBURSEMENT PER PRODUCT


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.ticker import FuncFormatter

loans_data = pd.read_excel('loans.xls')
grouped_data = loans_data.groupby('Product Name')['Loan Amount'].sum().reset_index()
grouped_data['Loan Amount (Millions)'] = grouped_data['Loan Amount'] / 1e6

sns.set(style="whitegrid")
plt.figure(figsize=(7, 4))
sns.lineplot(x='Product Name', y='Loan Amount (Millions)', data=grouped_data, color='blue')
plt.title('YTD Disbursement by Product Name')
plt.xlabel('Product Name')
plt.ylabel('Sum of Loan Amount (Millions)')
plt.xticks(rotation=45)  

formatter = FuncFormatter(lambda x, _: f'Kshs{x:.0f}M')
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()
```

MONTH TO MONTH TOTAL LOAN DISBURSEMENT

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

loans_data = pd.read_excel('loans.xls')
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

loans_data['Activation Month'] = pd.Categorical(loans_data['Activation Month'], categories=month_order, ordered=True)
grouped_data = loans_data.groupby('Activation Month')['Loan Amount'].sum().reset_index()
grouped_data['Loan Amount (Millions)'] = grouped_data['Loan Amount'] / 1e6
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.lineplot(x='Activation Month', y='Loan Amount (Millions)', data=grouped_data, color='blue')
plt.title('Month to Month total Loan disbursement')
plt.xlabel('Activation Month')
plt.ylabel('Sum of Loan Amount (Millions)')
plt


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

settlement_data = pd.read_excel('settlement.xls')
grouped_data = settlement_data.groupby('Closure Reason')['Principal-Balance'].sum().reset_index()
plt.figure(figsize=(7, 5))
plt.pie(grouped_data['Principal-Balance'], labels=grouped_data['Closure Reason'], autopct='%1.1f%%', startangle=90)
plt.title('Settlement Distribution based on Sum of Principal-Balance by Closure Reason')
plt.show()
```

YTD NUMBER OF LOANS DISBURSED PER PRODUCT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


loans_data = pd.read_excel('loans.xls')

Loan_count_per_product = loans_data['Product Name'].value_counts().reset_index()
Loan_count_per_product.columns = ['Product Name', 'Account ID']
Loan_count_per_product = Loan_count_per_product.sort_values(by='Account ID', ascending=False)
plt.figure(figsize=(7, 4))
sns.barplot(x='Account ID', y='Product Name', data=Loan_count_per_product, palette='viridis')
plt.title('Number of Loan Disbursed per Product')
plt.xlabel('Account ID')
plt.ylabel('Product Name')
plt.show()
```

YTD TOTAL LOAN SETTLEMENT PER PRODUCT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.ticker import FuncFormatter

settlement_data = pd.read_excel('settlement.xls')
grouped_data = settlement_data.groupby('Product Name')['Principal-Balance'].sum().reset_index()
grouped_data['Principal-Balance (Millions)'] = grouped_data['Principal-Balance'] / 1e6

sns.set(style="whitegrid")
plt.figure(figsize=(7, 4))
sns.barplot(x='Product Name', y='Principal-Balance (Millions)', data=grouped_data, color='blue')
plt.title('YTD Loan Settlement by Product Name')
plt.xlabel('Product Name')
plt.ylabel('Sum of Principal-Balance (Millions)')
plt.xticks(rotation=45)

formatter = FuncFormatter(lambda x, _: f'Kshs{x:.0f}M')
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()
```

Correlation between Count of Loans Disbursed and Sum of Loan Amount per Product 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

loans_data = pd.read_excel('loans.xls')
grouped_data = loans_data.groupby('Product Name').agg({'Loan Amount': 'sum', 'Account ID': 'count'}).reset_index()
grouped_data['Loan Amount (Millions)'] = grouped_data['Loan Amount'] / 1e6

plt.figure(figsize=(6, 4))
sns.scatterplot(x='Loan Amount (Millions)', y='Account ID', data=grouped_data, hue='Product Name', palette='viridis', s=100)

# Adding labels and title
plt.title('Correlation between Sum of Loan Amount and Count of Account ID per Product Name')
plt.ylabel('Count of Account ID')
plt.xlabel('Sum of Loan Amount (Millions)')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
